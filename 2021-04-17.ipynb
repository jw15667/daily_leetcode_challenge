```python
def test(rooms):
    for i in range(len(rooms[::-1])-1):

        last_room=len(rooms)-i-1
        #print(len(rooms)-i-1)
        #print(last_room, "(last_room)")
        new_rooms=rooms[0:last_room]+rooms[last_room+1:len(rooms)]
        #print(new_rooms, "(new_rooms)")
        found=any(last_room in sl for sl in new_rooms)
        if found==False:
            return found
        else:
            continue

    return found
```

```python
found=test(rooms)
found
```
